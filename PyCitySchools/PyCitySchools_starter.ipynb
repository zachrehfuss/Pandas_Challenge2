```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load 
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.
school_data_complete = pd.merge(student_data, school_data, how="left", on="school_name")
school_data_complete.head()
```

## District Summary

```python
# Display each Unique schools name
unique_schools = school_data_complete.loc[:, "school_name"].unique()

unique_schools
```

```python
#Calculate the number of unique schools
unique_schools_count = school_data_complete.loc[:, "school_name"].nunique()
unique_schools_count
```

```python
# Calculate the total number of students
student_count = school_data_complete.loc[:, "Student ID"].count()
student_count
```

```python
# Calculate the total budget
total_budget = school_data_complete.loc[:, "budget"].unique().sum()
total_budget
# ask question on how to make this into calculating only unique rows
```

```python
# Calculate the average (mean) math score
average_math_score = school_data_complete.loc[:, "math_score"]

results_math = average_math_score.mean()
results_math
```

```python
# Calculate the average (mean) reading score
average_reading_score = school_data_complete.loc[:, "reading_score"]
results_reading = average_reading_score.mean()
results_reading
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
passing_math_percentage
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)
passing_reading_count = school_data_complete[(school_data_complete["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(student_count) * 100
passing_reading_percentage
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
overall_passing_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
# The key metrics are; number of unique schools, total students, total budget, avg math score
# avg reading score, passing math %, passing read %, overall pas rate
district_summary = pd.DataFrame(
    {
        "Total Schools": [unique_schools_count],
        "Total Students": [student_count],
        "Total Budget": [total_budget],
        "Average Math Score": [results_math],
        "Average Reading Score": [results_reading],
        "Students Who Passed Math": [passing_math_percentage],
        "Students Who Passed Reading": [passing_reading_percentage],
        "Overall Passing Percentage": [overall_passing_rate]
        } 
)

# Formatting

district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)
district_summary["Average Math Score"] = district_summary["Average Math Score"].map("{:,.2f}".format)
district_summary["Average Reading Score"] = district_summary["Average Reading Score"].map("{:,.2f}".format)
district_summary["Students Who Passed Math"] = district_summary["Students Who Passed Math"].map("{:,.2f}".format)
district_summary["Students Who Passed Reading"] = district_summary["Students Who Passed Reading"].map("{:,.2f}".format)
district_summary["Overall Passing Percentage"] = district_summary["Overall Passing Percentage"].map("{:,.2f}".format)

# Display the DataFrame
district_summary
```

## Scores by School Spending

```python
# Establish the bins
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary for later aggregations
school_spending_df = per_school_summary.copy()
```

```python
# Use `pd.cut` on the per_school_capita Series from earlier to categorize per student spending based on the bins.
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df["Budget Per Student"], spending_bins, labels=labels,
    include_lowest=True)

# Convert Spending Ranges (Per Student) to a string
school_spending_df["Spending Ranges (Per Student)"] = school_spending_df["Spending Ranges (Per Student)"].astype(str)
school_spending_df
```

```python
#  Calculate averages for the desired columns.
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Passed Math Percent"].mean()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Passed Reading Percent"].mean()
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Overall Passing Percentage"].mean()
```

```python
# Assemble into DataFrame
spending_summary = pd.DataFrame({
    "Average Math Score" : spending_math_scores,
    "Average Reading Score" : spending_reading_scores,
    "Passed Math Percentage" : spending_passing_math,
    "Passed Reading Percentage" : spending_passing_reading,
    "Overall Passing Percentage" : overall_passing_spending
})

# Format the scores to 2 decimal points
spending_summary["Average Math Score"] = spending_summary["Average Math Score"].map("{:,.2f}".format)
spending_summary["Average Reading Score"] = spending_summary["Average Reading Score"].map("{:,.2f}".format)
spending_summary["Passed Math Percentage"] = spending_summary["Passed Math Percentage"].map("{:,.2f}".format)
spending_summary["Passed Reading Percentage"] = spending_summary["Passed Reading Percentage"].map("{:,.2f}".format)
spending_summary["Overall Passing Percentage"] = spending_summary["Overall Passing Percentage"].map("{:,.2f}".format)

# Display results
spending_summary
```
